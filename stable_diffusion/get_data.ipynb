```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
from torchvision.datasets import CocoCaptions
from PIL import Image
import random

# Path setup
image_dir = '/home/cesar/Downloads/mscoco/val2017/val2017'
ann_file = '/home/cesar/Downloads/mscoco/annotations_trainval2017/annotations/captions_val2017.json'

# Transform (optional, just for viewing)
transform = transforms.ToTensor()

# Dataset (use CocoCaptionsIgnoreMissing if needed)
dataset = CocoCaptions(root=image_dir, annFile=ann_file, transform=transform)

# Pick 9 random indices
indices = random.sample(range(len(dataset)), 9)

# Plot
fig, axs = plt.subplots(3, 3, figsize=(15, 15))

for ax, idx in zip(axs.flatten(), indices):
    image, captions = dataset[idx]

    print(image.size())
    
    # Convert tensor back to PIL image for display
    image = transforms.ToPILImage()(image)
    
    ax.imshow(image)
    ax.set_title(captions[0], fontsize=10)  # Use the first caption
    ax.axis('off')

plt.tight_layout()
plt.show()
```
